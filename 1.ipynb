```python
import numpy as np
from sklearn import svm , metrics
import matplotlib.pyplot as plt
```

```python

# Generate data for 2D function
# X = np.random.randn(400, 2)
# y = np.zeros(400)
# y[((X[:, 0] - 1) ** 2 + (X[:, 1] - 1) ** 2) < 0.5] = 1
# y[((X[:, 0] + 1) ** 2 + (X[:, 1] + 1) ** 2) < 1] = 1

# X = np.random.uniform(-np.pi, np.pi, (400, 2))
# y = np.sin(X[:, 0]) > X[:, 1]

X = np.random.randn(400, 2)
y = X[:, 0] > X[:, 1]

# from sklearn.model_selection import train_test_split
# x_train, x_test = train_test_split(x, test_size=0.2, random_state=42)
# y_train, y_test = train_test_split(y, test_size=0.2, random_state=42)



```

```python

# Train SVM model on data
model = svm.SVC(kernel='linear', C=1)
model.fit(X, y)

fig, ax = plt.subplots()
ax.scatter(X[y == 0, 0], X[y == 0, 1], c='red', cmap=plt.cm.Paired)
ax.scatter(X[y == 1, 0], X[y == 1, 1], c='blue', cmap=plt.cm.Paired)


plt.show()
```

```python

# Create meshgrid for plotting
xx, yy = np.meshgrid(np.linspace(-3, 3, 500), np.linspace(-3, 3, 500))
Z = model.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
```

```python

# Plot classification function and margin
fig, ax = plt.subplots()
ax.scatter(X[y == 0, 0], X[y == 0, 1], c='red', cmap=plt.cm.Paired)
ax.scatter(X[y == 1, 0], X[y == 1, 1], c='blue', cmap=plt.cm.Paired)
ax.contour(xx, yy, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
           linestyles=['--', '-', '--'])


# Set title and save plot
ax.set_title('sigmoid SVM on a simple function')
plt.savefig('sigmoid_simple.png')


# Show plot
plt.show()
```
