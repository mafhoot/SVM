```python
import numpy as np
from sklearn import svm
import matplotlib.pyplot as plt

# Generate data for 2D function
X = np.random.randn(200, 2)
y = np.logical_xor(X[:, 0] > 0, X[:, 1] > 0)
y = np.where(y, 1, -1)

# Train SVM model on data
model = svm.SVC(kernel='sigmoid', C=1)
model.fit(X, y)

# Create meshgrid for plotting
xx, yy = np.meshgrid(np.linspace(-3, 3, 500), np.linspace(-3, 3, 500))
Z = model.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot classification function and margin
fig, ax = plt.subplots()
ax.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)

# Plot decision boundary and margin
ax.contour(xx, yy, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
           linestyles=['--', '-', '--'])

# Set axis limits and remove background color
ax.set_xlim([-3, 3])
ax.set_ylim([-3, 3])
ax.set_facecolor('none')

# Set title and save plot
ax.set_title('sigmoid SVM on xor function')
plt.savefig('sigmoid_xor.png')

# Show plot
plt.show()
```
